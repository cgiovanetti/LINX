```python
%autoreload

import jax.numpy as jnp

import sys
sys.path.append('..')
```

```python

import matplotlib.pyplot as plt

import matplotlib.pylab as pylab

%matplotlib inline

from plot_params import params
pylab.rcParams.update(params)

cols_default = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
from linx.nuclear import NuclearRates
```

```python
PRIMAT_2018 = NuclearRates(nuclear_net='key_PRIMAT_2018')
PRIMAT_2022 = NuclearRates(nuclear_net='key_PRIMAT_2023')
YOF       = NuclearRates(nuclear_net='key_YOF')
```

```python
plt.figure()

rxn_ind = 11

T_ary = jnp.logspace(8., 9., num=50)

plt.loglog()

plt.plot(T_ary, YOF.reactions[rxn_ind].frwrd_rate_param(T_ary, 0.))
plt.plot(T_ary, PRIMAT_2022.reactions[rxn_ind].frwrd_rate_param(T_ary, 0.), '--')
plt.title(YOF.reactions[rxn_ind].name)

plt.fill_between(T_ary, YOF.reactions[rxn_ind].frwrd_rate_param(T_ary, -5.), YOF.reactions[rxn_ind].frwrd_rate_param(T_ary, 5.), alpha=0.2, color=cols_default[0])
plt.fill_between(T_ary, PRIMAT_2022.reactions[rxn_ind].frwrd_rate_param(T_ary, -5.), PRIMAT_2022.reactions[rxn_ind].frwrd_rate_param(T_ary, 5.), alpha=0.2, color=cols_default[1])
```

```python
plt.figure()

rxn_ind = 5

T_ary = jnp.logspace(8., 9., num=50)

plt.loglog()

plt.plot(T_ary, PRIMAT_2018.reactions[rxn_ind].frwrd_rate_param(T_ary, 0.))
plt.plot(T_ary, PRIMAT_2022.reactions[rxn_ind].frwrd_rate_param(T_ary, 0.), '--')
plt.title(PRIMAT_2018.reactions[rxn_ind].name)

plt.fill_between(T_ary, PRIMAT_2018.reactions[rxn_ind].frwrd_rate_param(T_ary, -5.), PRIMAT_2018.reactions[rxn_ind].frwrd_rate_param(T_ary, 5.), alpha=0.2, color=cols_default[0])
plt.fill_between(T_ary, PRIMAT_2022.reactions[rxn_ind].frwrd_rate_param(T_ary, -5.), PRIMAT_2022.reactions[rxn_ind].frwrd_rate_param(T_ary, 5.), alpha=0.2, color=cols_default[1])
```
