# Calculate Thermodynamics Varying $N_{\rm{eff}}$

The first step for any BBN calculation is to evaluate the thermodynamics as a function of time and scale factor, i.e. the energy densities and pressures of all the fluids. 

In LINX, we have implemented one such model: the $\Lambda$CDM + $N_\mathrm{eff}$ model, which is simply standard $\Lambda$CDM cosmology with the addition of an inert, always relativistic degree of freedom, with some initial energy density that then subsequently always redshifts as $\rho \propto a^{-4}$. 

These models are implemented in the class `thermo_model.ThermoModel`. First we initialize the model. The class instance can now be called directly, with the only model parameter of the model: $\Delta N_\mathrm{eff}$ (we need to define this somewhere). 

The result of this is a tuple of vectors, containing the time, the scalar factor, the *photon* energy density, the energy density of a single neutrino species (we do not distinguish between different neutrino species), the energy density of the inert species, the pressure of the inert species, and $N_\mathrm{eff}$. 

```python
# compile run

# get time, scale factor, photon energy density, neutrino energy density, extra energy density,
# given a delta Neff of 0:
t_vec, a_vec, rho_g_vec, rho_nu_vec, rho_extra_vec, P_extra_vec, Neff_vec = get_thermodynamics_Neff(0.)

# pass these quantities into GetAbundances to calculate abundances given this
# thermodynamic history
Yn, Yp, Yd, Yt, YHe3, Ya, YLi7, YBe7 = GetAbundances(
            jnp.array(rho_g_vec), # photon energy density
            jnp.array(rho_nu_vec), # neutrino energy density
            jnp.array(rho_extra_vec), # energy density of extra species
            jnp.array(rho_extra_vec)/3, # pressure of extra species
            t_vec=jnp.array(t_vec), # vector of times at which above energy densities are given
            a_vec=jnp.array(a_vec) # vector of scale factor at corresponding times
            )

print("N_eff is ",np.round(Neff_vec[-1],3))
print("Delta N_eff is ",3.045-np.round(Neff_vec[-1],3))
print("D/H is ",np.round(Yd/Yp*1e5,2))
print("Yp is ",np.round(4*Ya,3))
```

Note we pass a float to get_thermodynamics_Neff; this is to avoid recompiling the thermodynamics model later in the notebook where we evaluate with other floats.

We can observe the usual behavior for the energy densities of neutrinos and photons:

```python
fig = plt.figure(figsize=(6,4))
plt.plot(t_vec,rho_nu_vec/rho_g_vec)
ax = plt.gca()
ax.set_xscale('log')
plt.xlabel("Time (s)")
plt.ylabel(r"$\frac{\rho_{\nu}}{\rho_{\gamma}}$")
plt.show()
```

The input to get_thermodynamics_Neff() is not $\Delta N_{\rm{eff}}$ (though an input of 0 guarantees $\Delta N_{\rm{eff}}=0$).  The the actual change to $N_{\rm{eff}}$ is computed by introducing an energy density for additional inert species, with a coefficient given by the input to the function.  A larger coefficient will still result in a larger change to $N_{\rm{eff}}$.  Input can be positive or negative.

```python
# get thermodynamics for a nonzero delta Neff.  
t_vec, a_vec, rho_g_vec, rho_nu_vec, rho_extra_vec, P_extra_vec, Neff_vec = get_thermodynamics_Neff(4.)

# pass these quantities into GetAbundances to calculate abundances given this
# thermodynamic history
Yn, Yp, Yd, Yt, YHe3, Ya, YLi7, YBe7 = GetAbundances(
            jnp.array(rho_g_vec),
            jnp.array(rho_nu_vec), 
            jnp.array(rho_extra_vec),
            jnp.array(rho_extra_vec)/3,
            t_vec=jnp.array(t_vec),
            a_vec=jnp.array(a_vec)
            )

print("N_eff is ",np.round(Neff_vec[-1],3))
print("Delta N_eff is ",np.round(Neff_vec[-1],3)-3.045) # note this is not 4!
print("D/H is ",np.round(Yd/Yp*1e5,2))
print("Yp is ",np.round(4*Ya,3))
```

Now we have nonzero energy density of the "extra" species, computed by get_thermodynamics_Neff().  We can confirm that the species is inert by looking for $a^{-4}$ scaling:

```python
fig = plt.figure(figsize=(6,4))
plt.loglog(t_vec,rho_g_vec*a_vec**4,label=r"$\rho_{\gamma}$")
plt.loglog(t_vec,rho_nu_vec*a_vec**4,label=r"$\rho_{\nu}$")
plt.loglog(t_vec,rho_extra_vec*a_vec**4,label=r"$\rho_{\rm{extra}}$")
plt.xlabel("Time (s)")
plt.ylabel(r"Energy density$\times a^4$ (MeV$^4$)")
plt.legend()
plt.show()

# neutrino line should not be totally straight because of non-instantaneous neutrino 
# decoupling.  Meanwhile the inert species should be a perfectly flat line:
print("rho_nu_vec diff: ",rho_nu_vec[0]*a_vec[0]**4 - rho_nu_vec[-1]*a_vec[-1]**4)
print("rho_extra_vec diff: ",rho_extra_vec[0]*a_vec[0]**4 - rho_extra_vec[-1]*a_vec[-1]**4)
```
